```python
import pandas as pd
import pymysql
import sqlalchemy
from urllib import parse
from dotenv import load_dotenv

import os 
import sqlalchemy

# load .env
load_dotenv(".env")

db_name = os.environ.get('DB_NAME')
db_user = os.environ.get('DB_USER')
db_password = os.environ.get('DB_PASSWORD')
db_host = os.environ.get('DB_HOST')
db_port = os.environ.get('DB_PORT')

engine = sqlalchemy.create_engine(f"mysql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}")
```

```python
pd.options.display.max_columns = None
```

```python
df = pd.read_sql_table('real_items', engine)
df.head(1)
```

# type_id값이 25인 데이터 제거 ( 48개 )

```python
df[df['type_id'] == 25].shape
```

```python
df[df['type_id'] == 25]['wine_id'][:10]
```

```python
df2 = df[df['type_id'] != 25]
```

```python
df2
```

# Price 0인 값 제거 (남은 데이터 49058)
# 1754개 -> type_id가 25인 값을 제거해서 줄어들었음

```python
df2[df2['Price'] == 0].shape
```

```python
df3 = df2[df2['Price'] != 0]
df3.shape
```

# alcohol 수치가 0인 데이터 제거 (17032개 -> 남은 데이터 32026개)

```python
df3[df3['alcohol'] == 0].shape
```

```python
df4 = df3[df3['alcohol'] != 0]
df4.shape
```

# alcohol 30 초과되는 데이터 제거 (남은 데이터 32021개)

```python
df4[df4['alcohol'] > 30].shape
```

```python
df4[df4['alcohol'] > 30].index
```

```python
df5 = df4.drop(df4[df4['alcohol'] > 30].index)
df5.shape
```

# 와인 정보 중 음수 값 제거 (남은 데이터 32017개)

```python
df6 = df5[~((df5[['acidity_y', 'tannin', 'intensity', 'sweetness', 'body']] < 0).any(axis=1))]
df6.shape
```

# tannin 7.5 초과 데이터 제거 (남은 데이터 32008개)

```python
df6[df6['tannin'] > 7.5].shape
```

```python
df7 = df6[df6['tannin'] <= 7.5]
df7.shape
```

```python
df7
```

```python
df7.to_sql("final_items", engine)
```
